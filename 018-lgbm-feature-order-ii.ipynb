## blend01

```python
import matplotlib.pyplot as plt
```

```python
submission_blend = pd.read_csv("./result/Submission_HomeCredit_Blend.csv")
submission_mix02 = pd.read_csv("./result/submission-018-mix02.csv")

a = submission_mix02.TARGET.values.copy()
b = submission_blend.TARGET.rank().values.copy()
b /= b.max()

plt.scatter(a, b, alpha=.1)
plt.show()

bld = (a + b) / (a.max() + b.max())

submission_mix["TARGET"] = bld
submission_mix.to_csv("./result/submission-018-blend0.csv", index=False)
```

```python
submission_mix.head()
```

## blend02

```python
submission_lgbfs = pd.read_csv("./result/submission-018-lgbfs-rank11.csv")


a = submission_mix02.TARGET.values.copy()
b = submission_blend.TARGET.rank().values.copy()
b /= b.max()
c = submission_lgbfs.TARGET.rank().values.copy()
c /= c.max()

bld2 = (a + b + c) / (a.max() + b.max() + c.max())

submission_mix["TARGET"] = bld2
submission_mix.to_csv("./result/submission-018-blend01.csv", index=False)
```

```python
plt.scatter(bld, bld2, alpha=.1)
plt.show()
```

```python
a = submission_mix02.TARGET.values.copy()
b = submission_blend.TARGET.rank().values.copy()
b /= b.max()
c = submission_lgbfs.TARGET.rank().values.copy()
c /= c.max()

bld3 = (a + b + 2 * c) / (a.max() + b.max() + 2 * c.max())

submission_mix["TARGET"] = bld3
submission_mix.to_csv("./result/submission-018-blend02.csv", index=False)
```
